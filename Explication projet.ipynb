# Recommendation System - Part II
Model-Based Collaborative Filtering System

Objective:
Implement a model-based collaborative filtering system to recommend items based on user purchase history and similarity of ratings.

Data Preparation:

*   Selected a subset of the dataset (first 10,000 rows) to create a utility matrix.
*   Constructed a utility matrix, representing user-item preferences, where most values are unknown (sparse matrix).

Singular Value Decomposition (SVD):

*   Transposed the utility matrix to obtain a matrix where rows represent products and columns represent users.
*   Applied Truncated SVD to decompose the utility matrix into a lower-dimensional form (10 components).

Correlation Matrix: Calculated the correlation matrix based on the decomposed matrix, capturing relationships between products.

Isolating a Product:
Randomly chose a product (Product ID # 6117036094) and isolated it from the correlation matrix.

Correlation Calculation:
Determined the correlation of the chosen product with all other items based on ratings provided by customers who bought the same product.

Recommendation Generation:

*   Identified the top 10 highly correlated products as recommendations.
*   Removed the item already bought by the customer from the list of recommendations.
